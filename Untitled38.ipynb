```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spotify_main=pd.read_csv("tracks.csv")
spotify_main.head()
```

```python
spotify_main.isnull().sum()
```

```python
sorted_tracks=spotify_main.sort_values("popularity",ascending=False)
sorted_tracks.head(20)
```

```python
spotify_main.describe().transpose()
```

```python
most_popular=spotify_main[spotify_main["popularity"]>90].sort_values("popularity",ascending=False)
most_popular.head(10)
```

```python
spotify_main[["artists"]]
```

```python
spotify_main["duration"]=spotify_main["duration_ms"].apply(lambda x: round(x/1000))
spotify_main.sort_values("duration",ascending=False)
```

```python
corr_spotify=spotify_main.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
```

```python
plt.figure(figsize=(14,6))
corr_map=sns.heatmap(corr_spotify,cmap="inferno",annot=True)
corr_map.set_title("correlation spotify ")
corr_map.set_xticklabels(corr_map.get_xticklabels(),rotation=90)
plt.show()
```

```python
sample_df=spotify_main.sample(int(0.004*len(spotify_main)))
plt.figure(figsize=(13,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="red").set(title="correlation between energy and loudness")
```

```python
plt.figure(figsize=(13,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness",color="b").set(title="correlation between energy and loudness")
```

```python
total_duration=spotify_main.duration
fig,ax=plt.subplots(figsize=(10,4))
fig=plt.bar(spotify_main.release_date,spotify_main.duration,data=spotify_main).set(title=("year vs duration"))
plt.xticks(rotation=90)
```

```python
spotify_main["dates"]=spotify_main["release_date"]
spotify_main["dates"]=pd.to_datetime(spotify_main["dates"])
years=spotify_main["dates"].dt.year
plt.figure(figsize=(10,6))
plt.hist(years)
plt.title("number of songs for years")
plt.show()
```
